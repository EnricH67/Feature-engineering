Agafa un conjunt de dades de tema esportiu que t'agradi i normalitza els atributs categòrics en dummy. Normalitza els atributs numèrics amb StandardScaler.

__NOTA:on diu l'enunciat "Normalitza" hauria de dir "Estandaritza"__

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
from pandas import DataFrame
```

```python
import sklearn
from sklearn.preprocessing import StandardScaler
```

+ *Importo el meu data set de tenis*

```python
tenis_df = pd.read_csv('Match.csv')
tenis_df
```

+ *Passo la funció "StandardScaler()" a les variables numèriques del meu data set*

```python
trans = StandardScaler()
tenis2_df = trans.fit_transform(tenis_df[['avg_minutes_game', 'avg_seconds_point', 'avg_minutes_set', 'match_minutes']])
```

```python
tenis2_df
```

+ *I obtinc un resultat en forma de array.Per tant ho converteixo en un data frame amb la funció "DataFrame()"*

```python
tenis_stdscaler_df = DataFrame(tenis2_df)
```

```python
tenis_stdscaler_df
```

+ *I reanomeno les columnes amb un nou nom que m'orienti de que ja tinc dades standaritzades*

```python
tenis_stdscaler_df.rename(columns = {0:'std_avg_min_game', 1:'std_avg_sec_point', 2: 'std_avg_min_set', 3:'std_match_minutes' }, inplace = True)
tenis_stdscaler_df
```

+ *I comprovo amb ".describe()" que les dades tenen una mean molt propera a zero i una standard deviation = 1*

```python
tenis_stdscaler_df.describe()
```

+ *Ara genero les noves columnes amb les dummy variables extretes de les variables categòriques que tenia el meu data set*

  *I utilitzo els paràmetres "prefix" i "prefix_sep" per tenir de cop els noms de les columnes noves sense el nom de la columna 
   original ("tournament"), d'on he extret les diferents categories com a prefixe i sense caràcter separador (guió baix);
    que és com queden per defecte quan passes la funció "pd.get_dummies()"*

```python
tenis_dummies_df = pd.get_dummies(tenis_df, columns = ['tournament', 'round'], prefix = '', prefix_sep = '')
tenis_dummies_df
```

```python
tenis_dummies_df.drop(['avg_minutes_game', 'avg_seconds_point', 'avg_minutes_set', 'match_minutes'], axis = 1, inplace = True)
tenis_dummies_df
```

+ *Finalment faig un "pd.concat()" per tenir el data set final amb les variables numèriques standaritzades i les variables
   categòriques convertides en variables dummy, que he generat en dues fases abans* 

```python
final_tenis_df = pd.concat([tenis_dummies_df, tenis_stdscaler_df], axis = 1)
final_tenis_df
```
